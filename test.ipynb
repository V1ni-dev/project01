```python
import pandas as pd

df = pd.read_csv("pizzas.csv")
df
```

```python
df.plot(kind="scatter", x="diametro", y="preco")
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["diametro"]]
y = df[["preco"]]

modelo.fit(x,y)
```

```python
print(modelo.predict([[23]])[0][0])
```
